```python
from matplotlib.image import imread
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

def segment_image(image_path):
    # Read the image
    image = imread(image_path)

    # Reshape the image into a 2D array of pixels
    w, h, d = original_shape = tuple(image.shape)
    image_array = np.reshape(image, (w * h, d))

    # Fit KMeans clustering on the image
    kmeans = KMeans(n_clusters=2, random_state=0).fit(image_array)

    # Get labels and centroids
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # Create an array of duplicates of cluster_centers_rows based on labels
    segmented_image = centroids[labels]

    # Reshape the segmented image back to original shape
    segmented_image = segmented_image.reshape((w, h, d))

    # Return dictionary with segmented image and clusterer
    return {'segmented_image': segmented_image, 'clusterer': kmeans}

# Example usage:
result = segment_image("cloud.jpeg")
segmented_image = result['segmented_image']
clusterer = result['clusterer']

print(segmented_image, clusterer)

# Plot segmented image
plt.figure(figsize=(6, 6))
plt.title('Segmented Image')
plt.imshow(segmented_image)
plt.axis('off')
plt.show()
```

```python
from matplotlib.image import imread
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

def segment_image():
    # Read the image
    image = imread("cloud.jpeg")

    # Reshape the image into a 2D array of pixels
    w, h, d = original_shape = tuple(image.shape)
    image_array = np.reshape(image, (w * h, d))

    # Fit KMeans clustering on the image
    kmeans = KMeans(n_clusters=2, random_state=0).fit(image_array)

    # Get labels and centroids
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # Create the segmented image based on the labels and centroids
    segmented_image = centroids[labels].astype(int)  #convert float to int

    # Reshape the segmented image back to original shape
    segmented_image = segmented_image.reshape((w, h, d))

    # Return dictionary with segmented image and clusterer
    return {'segmented_image': segmented_image, 'clusterer': kmeans}

# Example usage:
result = segment_image()
segmented_image = result['segmented_image']
clusterer = result['clusterer']
print(segmented_image, clusterer)

# Plot segmented image
plt.figure(figsize=(6, 6))
plt.title('Segmented Image')
plt.imshow(segmented_image.astype(np.uint8))  # Convert to uint8 for proper display
plt.axis('off')
plt.show()
```
